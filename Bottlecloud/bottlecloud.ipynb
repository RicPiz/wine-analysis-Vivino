```python
import numpy as np
from PIL import Image
from wordcloud import WordCloud
```

```python
# For red wine use decanter.png

wine_mask = np.array(Image.open("wine-bottle-transparent.png"))
wine_mask
```

```python
wine_mask[0]
```

```python
for i in wine_mask:
    i[i == 0] = 255

wine_mask
```

```python
with open("white_reviews_cloud.txt", "r") as f:
    text = f.read()
```

```python
words = ''

for word in text:
    word = str(word)
    words += "".join(word)+""
```

```python
import matplotlib.pyplot as plt

# Create a word cloud image
wc = WordCloud(background_color='black', max_words=1000, min_font_size=2, mask=wine_mask,
               colormap='YlOrBr', relative_scaling=0, random_state=42, width=1080, height=1080)

# Generate a wordcloud
wc.generate(words)

# store to file
wc.to_file("bottlecloud_w.png")

with open("white_wines_bottlecloud_white.svg", "w") as text_file:
    text_file.write(wc.to_svg())

# show
plt.figure(figsize=[10,18], dpi=300)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```
